## Loading/analyzing files

#### Create randomized 2D data, write to file, read from file 

```python
num = 20
data = zip(np.random.rand(num),np.random.rand(num))
fileName = "test"
np.savetxt(fileName, data)
data = np.loadtxt(fileName)
```

#### A line plot 

```python
# create new data set 
data2 = data + 0.5*np.reshape(np.random.rand(num*2),[num,2])

plt.plot(data[:,0],data[:,1], label="data1")
plt.plot(data2[:,0],data2[:,1], label="data2")
plt.legend(loc=0)
plt.xlabel("xs")
plt.ylabel("ys")
plt.gcf().savefig("myfile.png")
```

#### Scatter

```python
plt.scatter(data[:,0],data[:,1], facecolor='r',label="data1")
plt.scatter(data2[:,0],data2[:,1], facecolor='w',label="data2")
```

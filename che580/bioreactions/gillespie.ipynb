## Gillespie simulator

This file contains a Gillespie simulator for the reaction
Borrowed from Markus Dittrich

<code>
   A  -> B  [k1]
   B -> A   [k2]
</code>

http://staff.psc.edu/dittrich/PGSS/pdfs/Gillespie_1977.pdf

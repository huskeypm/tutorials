```python
import numpy as np
%pylab inline
import matplotlib.pylab as plt
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
Es = np.linspace(-5,5,100)
kT= 0.59 # kcal/mol
N = np.exp(-Es/kT)
plt.plot(Es,N, label="$N=g\;e^{-E/kT}$")
plt.ylabel("N (population)")
plt.legend(loc=0)
plt.xlabel("E[kcal/mol]")
plt.title("Boltzmann distribution")
```

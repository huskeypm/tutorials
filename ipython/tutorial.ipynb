## Basic notebook

This notebook assumes you have a two-column text file (xy values).  

```python
fileName = "data.txt"

createTextFile = True
# create randomized data if datafile doesnt exist
if createTextFile:
    ts = np.linspace(0,1000,1001)
    ys1 = (ts/1000.)**2 + 0.1*np.random.randn( np.shape(ts)[0] )# parabola w noise
    ys2 = np.exp(-ts/100.) + 0.1*np.random.randn( np.shape(ts)[0] ) # exp w noise 
    data = np.zeros([np.shape(ts)[0],3])
    
    # save
    data[:,0] = ts
    data[:,1] = ys1
    data[:,2] = ys2
    np.savetxt(fileName,data)  
    
# load existing data file     
else:
    data = np.loadtxt(fileName)
```

```python
plt.plot(data[:,0],data[:,1],label="ys1")
plt.plot(data[:,0],data[:,2],label="ys2")
plt.xlabel("time [ms]")
plt.legend( loc=0 )

# save file 
plt.gcf().savefig("x.png")
```
